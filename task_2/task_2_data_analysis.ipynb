```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import root_mean_squared_error
import matplotlib.pyplot as plt
```

# Model train

### Linear regression (Base model)

```python
X = df.drop(columns='target')
Y = df['target']
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
```

```python
linear_regression_model = LinearRegression()
linear_regression_model.fit(x_train, y_train)

y_predicted = linear_regression_model.predict(x_test)
root_mean_squared_error(y_test, y_predicted)
```

### Decision tree regressor

```python
tree_regressor = DecisionTreeRegressor(random_state=42)
tree_regressor.fit(x_train, y_train)

y_predicted = tree_regressor.predict(x_test)
root_mean_squared_error(y_test, y_predicted)
```

### Random forest regressor

```python
forest_regressor = RandomForestRegressor(n_estimators=30, random_state=42, max_depth=15)
forest_regressor.fit(x_train, y_train)

y_predicted = forest_regressor.predict(x_test)
root_mean_squared_error(y_test, y_predicted)
```

As we can see decision tree regressor shows good results with low time consuming training process. As for now we can stop on this model.
